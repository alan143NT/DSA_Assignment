```python
df = df.drop(['player_id','name'],axis=1)
```

```python
df.head(10)
```

```python

# Convert height into inches

def convert_height(hgt):
    feet, inches = hgt.split("'")
    total_inches = int(feet) * 12 + int(inches.replace('"', ''))
    return total_inches

df['height'] = df['height'].apply(convert_height)
```

```python
# Remove the unit "lbs" from the data

df['weight'] = df['weight'].str.replace('lbs', '')
```

```python
# convert weight to Float type
df['weight'] = pd.to_numeric(df['weight'], errors='coerce').astype('Float64')
```

### Handling Null Values

```python
numerical = ['age', 'height', 'weight', 'ball_controlling_skills',
       'body_reflexes', 'body_balance', 'jumping_skills',
       'penalties_conversion_rate', 'mental_strength', 'goalkeeping_skills',
       'defending_skills', 'passing_skills', 'dribbling_skills',
       'shot_accuracy', 'body_strength_stamina', 'max_running_speed',
       'behaviour_rating', 'matches_played', 'fitness_rating',
       'trophies_won', 'years_of_experience',
       'no_of_disqualifications']
categorical = ['gender','strong_foot','coaching','dedication_level']
```

```python
for i in categorical:
    df[i]=df[i].fillna(df[i].mode()[0])
```

```python
for i in numerical:
    df[i]=df[i].fillna(df[i].median())
```

### Encoding 

```python
# Label encoding and one hot encoding
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
le = LabelEncoder()
x= df[['dedication_level','country']]

for i in x:
    df[i] = le.fit_transform(df[i])
```

```python

categ = ['gender', 'strong_foot', 'coaching']
ohe = OneHotEncoder()

# Reshape and transform each categorical feature
for i in categ:
    transformed_data = ohe.fit_transform(df[i].values.reshape(-1, 1))
    df_transformed = pd.DataFrame(transformed_data.toarray(), columns=ohe.get_feature_names_out([i]))
    df = pd.concat([df, df_transformed], axis=1)

# Drop the original categorical columns
df.drop(categ, axis=1, inplace=True)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
for i in numerical:
    numb=np.array(df[i])
    numb=numb.reshape(-1, 1)
    df[i] = scaler.fit_transform(numb)
```

### Model Selection

```python
x= df.drop(['selection'],axis=1)
```

```python
y= df['selection']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=20,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
clf = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the model on the training data
dt_train = clf.fit(x_train, y_train)
```

```python
# Predict on the test set
y_pred = clf.predict(x_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
```

```python
print("Accuracy:", accuracy)
```
